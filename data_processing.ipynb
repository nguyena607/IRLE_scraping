```python
import json
import pandas as pd

file_path = 'prices.jsonl'

with open(file_path, 'r') as file:
    data = json.load(file)

# Initialize a list to store the data
df_data = []

# Iterate through the data and extract the information
for search_location, restaurants in data.items():
    for restaurant_name, menu in restaurants.items():
        if not menu:  # if menu is empty, i.e., the restaurant has no items
            df_data.append({
                "search_location": search_location,
                "restaurant_name": restaurant_name,
                "item": None,
                "price": None,
                "restaurant_address": None
            })
        else:

            restaurant_address = menu['location']
            for item, price in menu.items():
                df_data.append({
                    "search_location": search_location,
                    "restaurant_name": restaurant_name,
                    "item": item,
                    "price": price,
                    "restaurant_address": restaurant_address
                })

# Convert the list to a DataFrame
df = pd.DataFrame(df_data)
```

```python
df.sample(100)
```

```python
df.to_csv('fast_food_restaurants_menu.csv', index=False) 
```
